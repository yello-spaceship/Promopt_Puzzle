```python
# 2. 라이브러리 임포트
import os
import json
import torch
import timm
import pandas as pd
from PIL import Image
from tqdm import tqdm
import torchvision.transforms as T
from diffusers import StableDiffusionPipeline
from transformers import CLIPProcessor, CLIPModel
```

```python
# 3. 장치 설정
device = "cuda" if torch.cuda.is_available() else "cpu"

# 4. 모델 로드
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16 if device=="cuda" else torch.float32
).to(device)

clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

aesthetic_model = timm.create_model('vit_base_patch16_224', pretrained=True, num_classes=1).to(device)
aesthetic_model.eval()
aesthetic_transform = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize([0.5], [0.5]),
])
```

```python
# 5. 평가 함수
def compute_clip_score(image: Image.Image, text: str):
    inputs = clip_processor(text=[text], images=image, truncation=True,return_tensors="pt", padding=True).to(device)
    with torch.no_grad():
        outputs = clip_model(**inputs)
        image_emb = outputs.image_embeds / outputs.image_embeds.norm(p=2, dim=-1, keepdim=True)
        text_emb = outputs.text_embeds / outputs.text_embeds.norm(p=2, dim=-1, keepdim=True)
        return (image_emb @ text_emb.T).item()

def compute_aesthetic_score(image: Image.Image):
    tensor = aesthetic_transform(image).unsqueeze(0).to(device)
    with torch.no_grad():
        return aesthetic_model(tensor).squeeze().item()

def compute_clip_embedding(image: Image.Image):
    inputs = clip_processor(images=image, return_tensors="pt").to(device)
    with torch.no_grad():
        image_features = clip_model.get_image_features(**inputs)
        return image_features / image_features.norm(p=2, dim=-1, keepdim=True)
from transformers import BlipProcessor, BlipForConditionalGeneration
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

def compute_blip_score(image, prompt):
    inputs = blip_processor(image, return_tensors="pt").to(device)
    generated_ids = blip_model.generate(**inputs)
    caption = blip_processor.decode(generated_ids[0], skip_special_tokens=True)

    # 간단한 유사도: 정답 문장(prompt)과 생성 caption의 BLEU/ROUGE 등 넣어도 되고,
    # 임시로 cosine similarity 또는 단순 string overlap 사용 가능
    from sentence_transformers import SentenceTransformer, util
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    emb1 = model.encode(prompt, convert_to_tensor=True)
    emb2 = model.encode(caption, convert_to_tensor=True)
    return float(util.cos_sim(emb1, emb2)[0][0])
```

```python
import os
import json
import pandas as pd
from tqdm import tqdm
from PIL import Image

# 이미지 저장 폴더 생성
os.makedirs("images", exist_ok=True)

results = []

cnt = 0
for item in tqdm(prompt_data):
    if cnt >= 100:
        break
    cnt += 1

    prompt_id = item["promptId"]
    base_prompt = item["original_prompt"]
    painter_prompt = item["director_prompt"]

    try:
        # 이미지 생성
        base_image = pipe(base_prompt).images[0]
        painter_image = pipe(painter_prompt).images[0]

        # 저장
        base_path = f"images/base_{prompt_id}.png"
        painter_path = f"images/eng_{prompt_id}.png"
        base_image.save(base_path)
        painter_image.save(painter_path)

        # 평가 - CLIP, Aesthetic, BLIP
        #base_clip = compute_clip_score(base_image, base_prompt)
        base_aesthetic = compute_aesthetic_score(base_image)
        base_blip = compute_blip_score(base_image, base_prompt)

        #painter_clip = compute_clip_score(painter_image, painter_prompt)
        painter_aesthetic = compute_aesthetic_score(painter_image)
        painter_blip = compute_blip_score(painter_image, painter_prompt)

        # 결과 저장
        results.append({
            "promptId": prompt_id,
            "base_prompt": base_prompt,
            "director_prompt": painter_prompt,
            "base_aesthetic": round(base_aesthetic, 4),
            "base_blip": round(base_blip, 4),
            "director_aesthetic": round(painter_aesthetic, 4),
            "director_blip": round(painter_blip, 4),
            "base_path": base_path,
            "director_path": painter_path
        })

        # 출력
        print(f"\n🧾 Prompt ID: {prompt_id}")
        print(f"▶️ Base Prompt:\n{base_prompt}")
        print(f"🎨 Engineer Prompt:\n{painter_prompt}")
        print("📊 평가 결과:")
        print(f"[BASE]    CLIP: {base_clip:.4f}, Aesthetic: {base_aesthetic:.4f}, BLIP: {base_blip:.4f}")
        print(f"[ENGINEER] CLIP: {painter_clip:.4f}, Aesthetic: {painter_aesthetic:.4f}, BLIP: {painter_blip:.4f}")

    except Exception as e:
        print(f"❌ Error with promptId {prompt_id}: {e}")
        continue

# 전체 결과 CSV 저장
df = pd.DataFrame(results)
df.to_csv("prompt_comparison_results_director.csv", index=False)
print("\n✅ 결과 CSV 저장 완료: prompt_comparison_results.csv")
```

```python
# ✅ Google Colab 기반 Persona Prompt Fusion 이미지 생성 파이프라인

from google.colab import drive
import os
import json
import torch
from PIL import Image
from tqdm import tqdm
from transformers import BlipProcessor, BlipForConditionalGeneration
from diffusers import StableDiffusionPipeline

# Mount Google Drive
drive.mount('/content/drive')

# Base path 설정
BASE_DIR = "/content/drive/MyDrive/fusion_project"
os.makedirs(BASE_DIR + "/outputs/images", exist_ok=True)
os.makedirs(BASE_DIR + "/outputs/logs", exist_ok=True)
os.makedirs(BASE_DIR + "/checkpoints", exist_ok=True)

# STEP 1: 데이터 로드
with open("/content/drive/MyDrive/fusion_project/train.json") as f:
    train_data = json.load(f)

# STEP 2: 모델 로딩
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").eval().cuda()
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16).to("cuda")

# STEP 3: 임베딩 추출 함수
@torch.no_grad()
def get_blip_text_embedding(prompt):
    inputs = processor(text=prompt, return_tensors="pt").to("cuda")
    return blip_model.text_encoder(**inputs).last_hidden_state.mean(dim=1)  # [1, 768]

# STEP 4: BLIP Score 계산
@torch.no_grad()
def compute_blip_score(image: Image, prompt: str):
    processor_itm = BlipProcessor.from_pretrained("Salesforce/blip-itm-base-coco")
    model_itm = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-itm-base-coco").cuda().eval()
    inputs = processor_itm(image, prompt, return_tensors="pt").to("cuda")
    output = model_itm(**inputs)
    score = torch.sigmoid(output.logits_per_text[0][0]).item()
    return score

# STEP 5: 퓨전 함수
def fuse_embeddings(embeddings, weights):
    return sum(w * e for w, e in zip(weights, embeddings)) / sum(weights)

def select_top2(persona_scores, base_score, delta=0.04):
    filtered = {k: v for k, v in persona_scores.items() if abs(v - base_score) <= delta and v > base_score}
    top2 = sorted(filtered.items(), key=lambda x: x[1], reverse=True)[:2]
    return [k for k, _ in top2]

# STEP 6: 실행 루프
persona_keys = ["painter", "poet", "minimal", "engineer", "director"]
results = []

for item in tqdm(train_data[:30]):  # 예시: 30개만 실행
    pid = item["promptId"]
    base_prompt = item["original_prompt"]

    # 1. base 이미지/점수
    base_image = pipe(base_prompt).images[0]
    base_image_path = f"{BASE_DIR}/outputs/images/{pid}_base.png"
    base_image.save(base_image_path)
    base_score = compute_blip_score(base_image, base_prompt)

    # 2. persona 이미지 및 점수
    persona_scores = {}
    persona_embeddings = {}
    for persona in persona_keys:
        prompt = item.get(persona)
        if not prompt:
            continue
        image = pipe(prompt).images[0]
        score = compute_blip_score(image, prompt)
        persona_scores[persona] = score
        persona_embeddings[persona] = get_blip_text_embedding(prompt)
        image.save(f"{BASE_DIR}/outputs/images/{pid}_{persona}.png")

    # 3. top2 선택 및 퓨전
    selected = select_top2(persona_scores, base_score)
    if not selected:
        continue
    embeddings = [persona_embeddings[p] for p in selected]
    fused_emb = fuse_embeddings(embeddings, [1.0] * len(embeddings))

    # 4. 프롬프트로 되돌리지는 않고, 프롬프트 concat으로 대체
    fused_prompt = " ".join([item[p] for p in selected])
    fused_image = pipe(fused_prompt).images[0]
    fused_path = f"{BASE_DIR}/outputs/images/{pid}_fused.png"
    fused_image.save(fused_path)
    fused_score = compute_blip_score(fused_image, fused_prompt)

    results.append({
        "promptId": pid,
        "base_score": round(base_score, 4),
        "fused_score": round(fused_score, 4),
        "selected_personas": selected,
        "fused_prompt": fused_prompt
    })

# STEP 7: 로그 저장
with open(f"{BASE_DIR}/outputs/logs/fusion_scores.json", "w") as f:
    json.dump(results, f, indent=2)
```
